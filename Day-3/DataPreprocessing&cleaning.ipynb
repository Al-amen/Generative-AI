### HTML tags removal


```python
import re 
def remove_html_tags(text):
    pattern = re.compile(r'<.*?>')
    return re.sub(pattern, '', text)
```

```python
text = "<html><body><p> Movie 1</p><p> Actor - Aamir Khan</p><p> Click here to <a href='http://google.com'>download</a></p></body></html>"
remove_html_tags(text)
```

```python
df['review'] = df['review'].apply(remove_html_tags)
```

### Remove url

```python
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return re.sub(pattern, '', text)
```

```python
text1 = 'Check out my youtube https://www.youtube.com/dswithbappy dswithbappy'
text2 = 'Check out my linkedin https://www.linkedin.com/in/boktiarahmed73/'
text3 = 'Google search here www.google.com'
text4 = 'For data click https://www.kaggle.com/'
```

```python
remove_url(text1)
```

```python
df['review'] = df['review'].apply(remove_url)
```

### Punctuation Removal

```python
import string, time
string.punctuation
```

```python
exlude_punctuation = string.punctuation
exlude_punctuation
```

```python
def remove_punctuation(text):
    for char in exlude_punctuation:
        text = text.replace(char, '')
    return text
```

```python
text = 'string. With. Punctuation?&@'
```

```python
remove_punctuation(text)
```

```python
start = time.time()
print(remove_punctuation(text))
time1 = time.time() - start
print(time1*50000)
```

```python
def remove_punctuation1(text):
    return text.translate(str.maketrans('', '', exlude_punctuation))
```

```python
start = time.time()
remove_punctuation1(text)
time2 = time.time() - start
print(time2*50000)
```

### Chat Conversion handle

```python
chat_words = {
    'AFAIK':'As Far As I Know',
    'AFK':'Away From Keyboard',
    'ASAP':'As Soon As Possible'
}


{
    "FYI": "For Your Information",
    "ASAP": "As Soon As Possible",
    "BRB": "Be Right Back",
    "BTW": "By The Way",
    "OMG": "Oh My God",
    "IMO": "In My Opinion",
    "LOL": "Laugh Out Loud",
    "TTYL": "Talk To You Later",
    "GTG": "Got To Go",
    "TTYT": "Talk To You Tomorrow",
    "IDK": "I Don't Know",
    "TMI": "Too Much Information",
    "IMHO": "In My Humble Opinion",
    "ICYMI": "In Case You Missed It",
    "AFAIK": "As Far As I Know",
    "BTW": "By The Way",
    "FAQ": "Frequently Asked Questions",
    "TGIF": "Thank God It's Friday",
    "FYA": "For Your Action",
    "ICYMI": "In Case You Missed It",
}
```

```python
def chat_conversion(text):
    new_text = []
    for word in text.split():
        if word.upper() in chat_words:
            new_text.append(chat_words[word.upper()])
        else:
            new_text.append(word)
    return " ".join(new_text)
```

```python
chat_conversion('Do this work ASAP')
```

### Stopwords
# stopwords are those words , does not carry strong meaning

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word not in stopwords.words('english'):
            new_text.append(word)
    return " ".join(new_text)
```

```python
remove_stopwords('probably my all-time favorite movie, a story of selflessness, sacrifice and dedication to a noble cause, but it\'s not preachy or boring. it just never gets old, despite my having seen it some 15 or more times')
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

#### Remove emoji handle

```python
import re 
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
remove_emoji("Loved the movie. It was üòòüòò")
```

## Stemmer 

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(test):
    return " ".join([ps.stem(word) for word in test.split()])
```

```python
sample = "walk walks walking walked"
stem_words(sample)
```

### Lemmatization

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
punctuations="?:!.,;"
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)

sentence_words
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_words:
    print ("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word,pos='v')))
```
